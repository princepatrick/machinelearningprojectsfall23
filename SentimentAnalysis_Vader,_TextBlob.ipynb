```python
import pandas as pd

cellphone_data = pd.read_json("/content/Cell_Phones_and_Accessories_5.json", lines=True)

cellphone_data.head(5)
```

```python
def drop_redundant_reviews(df):
  new_df = df.drop_duplicates(subset=['reviewerID'])
  return new_df
```

```python
df = drop_redundant_reviews(cellphone_data)
```

```python
df = df.drop(columns=['helpful', 'unixReviewTime'])
```

Using TextBlob to check the polarity and subjectivity. The Polarity explains the emotion of the review whether it is a positive or a negative review and the subjectivity explains whether it is a personal opinion or a factual statement.

Text Blob

```python
from textblob import TextBlob

for row in range(0,20):
  revText = df['reviewText'].iloc[row]
  print(f"The polarity for {revText} is {TextBlob(revText).sentiment.polarity}")
```

Subjectivity

```python
for row in range(0,20):
  revText = df['reviewText'].iloc[row]
  print(f"The subjectivity for {revText} is {TextBlob(revText).sentiment.subjectivity}")
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

print("Vader Sentiment Analysis")

for i in range(0,20):
  print(f'The polarity scores for {df["reviewText"].iloc[i]} is {sia.polarity_scores(df["reviewText"].iloc[i])}')
```
